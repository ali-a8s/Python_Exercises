```python
# Exercise 10: Merge two lists using the following condition
# odd numbers from the first list and even numbers from the second list.

lst1 = [10, 20, 25, 30, 35]
lst2 = [40, 45, 60, 75, 90]

res = []

for i in lst1:
  if i % 2 != 0:
    res.append(i)

for j in lst2:
  if j % 2 == 0:
    res.append(j)

print(res)
```

```python
# Exercise 18: Check if a given year is a leap year

def is_leap_year(year):
  print("Year", year)
  if year % 4 == 0:
    if year % 100 == 0:
      if year % 400 == 0:
        return True
      else:
        return False
    else:
      return True
  else:
    return False

# check
print(is_leap_year(2020))
print(is_leap_year(2025))
```

```python
# Exercise: 19: Print Alternate Prime Numbers till 20

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

n = 20
primes = []
for num in range(2, n + 1):
    if is_prime(num):
        primes.append(num)

print(f'All prime numbers from 1 to {n}: {primes}')

print(f'Alternate prime numbers from 1 to {n}:')

for i in range(0, len(primes), 2):
    print(primes[i])
```

```python
# Exercise 20: Print Reverse Number Pattern

def print_pattern(rows):
  for i in range(1, rows + 1):
    for j in range(rows - i + 1):
      print(i, end=" ")
    print()

print_pattern(5)
```

```python
# Exercise 21: Check if a user-entered string contains any digits using a for loop

def contains_digits(text):
  for char in text:
    if '0' <= char <= '9':
      return True
  return False

# Get input from the user
user_string = input("Enter a string: ")

if contains_digits(user_string):
  print("The string contains at least one digit.")
else:
  print("The string does not contain any digits.")
```

```python
# Exercise 22: Capitalize the first letter of each word in a string

def capt_words(txt):
  ls = txt.split()
  ls = [word.capitalize() for word in ls]
  return " ".join(ls)

str1 = "pynative.com is for python lovers"

capitalized_string = capt_words(str1)
print("Capitalized string:", capitalized_string)
```

```python
# Exercise 23: Create a simple countdown timer using a while loop.
import time

def count_down(seconds):
  while seconds > 0:
    print(f"Time remaining: {seconds} seconds")
    time.sleep(1)
    seconds -= 1
  print("time's up")

count_down(10)
```
